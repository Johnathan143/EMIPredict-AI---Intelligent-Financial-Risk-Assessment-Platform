```python
import pandas as pd

df = pd.read_csv("emi_cleaned.csv")

df.head()
```

## Total Expenses

```python
df["total_expenses"] = (
    df["monthly_rent"]
    + df["school_fees"]
    + df["college_fees"]
    + df["travel_expenses"]
    + df["groceries_utilities"]
    + df["other_monthly_expenses"]
)
```

## Debt-to-Income Ratio

```python
df["debt_to_income_ratio"] = (
    df["current_emi_amount"]
    / df["monthly_salary"]
)
```

## Expense Ratio

```python
df["expense_ratio"] = (
    df["total_expenses"]
    / df["monthly_salary"]
)
```

## Savings Ratio

```python
df["savings_ratio"] = (
    (df["bank_balance"] + df["emergency_fund"])
    / df["monthly_salary"]
)
```

## Financial Cushion

```python
df["financial_cushion"] = (
    df["bank_balance"]
    + df["emergency_fund"]
)
```

## Employment Stability

```python
df["employment_stability"] = (
    df["years_of_employment"]
    * df["credit_score"]
)
```

## Loan Burden Ratio

```python
df["loan_burden_ratio"] = (
    df["requested_amount"]
    / df["monthly_salary"]
)
```

## Available Income

```python
df["available_income"] = (
    df["monthly_salary"]
    - df["total_expenses"]
    - df["current_emi_amount"]
)
```

```python
new_features = [
    "total_expenses",
    "debt_to_income_ratio",
    "expense_ratio",
    "savings_ratio",
    "financial_cushion",
    "employment_stability",
    "loan_burden_ratio",
    "available_income"
]

df[new_features].head()
```
